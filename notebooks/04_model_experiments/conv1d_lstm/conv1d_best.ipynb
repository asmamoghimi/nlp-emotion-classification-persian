```python
import stanza 

stanza.download('fa')

nlp = stanza.Pipeline("fa", processors="tokenize,pos")
```

```python
from tqdm import tqdm

pos_tags = []
batch_size = 100   

texts = df["text"].fillna("").astype(str).tolist()

for i in tqdm(range(0, len(texts), batch_size), desc="Processing batches"):
    batch = texts[i:i+batch_size]
    for text in batch:
        doc = nlp(text)   
        tags = [word.upos for sent in doc.sentences for word in sent.words]
        pos_tags.append(tags)

df["pos_tags"] = pos_tags
```

```python
df.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
all_tags = [tag for tags in df["pos_tags"] for tag in tags]
pos_encoder = LabelEncoder()
pos_encoder.fit(all_tags)
```

```python
pos_sequences = [pos_encoder.transform(tags) for tags in df["pos_tags"]]
```

```python
max_len = 50
X_pos = pad_sequences(pos_sequences, maxlen=max_len, padding="post")

print("X_pos shape:", X_pos.shape)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

df["text"] = df["text"].fillna("").astype(str)

tokenizer = Tokenizer(num_words=20000, oov_token="<OOV>")
tokenizer.fit_on_texts(df["text"])

X_words = tokenizer.texts_to_sequences(df["text"])
X_words = pad_sequences(X_words, maxlen=max_len, padding="post")

print("X_words shape:", X_words.shape)
```

```python
from tensorflow.keras.utils import to_categorical

num_classes = len(df["label_encoded"].unique())
y = to_categorical(df["label_encoded"], num_classes=num_classes)

print("y shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

Xw_train, Xw_val, Xp_train, Xp_val, y_train, y_val = train_test_split(
    X_words, X_pos, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Conv1D, GlobalMaxPooling1D, Dense, Dropout, Concatenate
from tensorflow.keras.regularizers import l2
```

```python
# Word input
word_input = Input(shape=(max_len,), name="word_input")
word_emb = Embedding(input_dim=20000, output_dim=128, input_length=max_len)(word_input)
word_conv = Conv1D(128, 5, activation="relu")(word_emb)
word_pool = GlobalMaxPooling1D()(word_conv)

# POS input
pos_input = Input(shape=(max_len,), name="pos_input")
pos_emb = Embedding(input_dim=len(pos_encoder.classes_)+1, output_dim=16, input_length=max_len)(pos_input)
pos_conv = Conv1D(64, 3, activation="relu")(pos_emb)
pos_pool = GlobalMaxPooling1D()(pos_conv)

# Merge both
merged = Concatenate()([word_pool, pos_pool])
dense = Dense(64, activation="relu", kernel_regularizer=l2(0.001))(merged)
drop = Dropout(0.6)(dense)
output = Dense(num_classes, activation="softmax")(drop)

model = Model(inputs=[word_input, pos_input], outputs=output)
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

# Compute class weights
class_weights = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(df["label_encoded"]),
    y=df["label_encoded"]
)
class_weights = dict(enumerate(class_weights))

early_stop = EarlyStopping(monitor="val_loss", patience=3, restore_best_weights=True)

history = model.fit(
    [Xw_train, Xp_train], y_train,
    batch_size=128,
    epochs=30,
    validation_data=([Xw_val, Xp_val], y_val),
    class_weight=class_weights,
    callbacks=[early_stop]
)
```

```python
train_loss, train_acc = model.evaluate([Xw_train, Xp_train], y_train, verbose=0)
val_loss, val_acc = model.evaluate([Xw_val, Xp_val], y_val, verbose=0)

print(f"Train accuracy: {train_acc:.3f}, Validation accuracy: {val_acc:.3f}")
```
